```python
from Classes.GX_BERT import projTransformer
from tensorflow import keras
import tensorflow as tf
import numpy as np
import pandas as pd

%load_ext tensorboard
```

# Prepare Dataset

```python
newDataset1_0 = pd.read_hdf("CTB/CTB_128k_slack_0.h5", key="df", mode="r")
newDataset1_1 = pd.read_hdf("CTB/CTB_128k_slack_1.h5", key="df", mode="r")
newDataset1_2 = pd.read_hdf("CTB/CTB_128k_slack_2.h5", key="df", mode="r")
newDataset1_3 = pd.read_hdf("CTB/CTB_128k_slack_3.h5", key="df", mode="r")
newDataset1_4 = pd.read_hdf("CTB/CTB_128k_slack_4.h5", key="df", mode="r")
newDataset1_5 = pd.read_hdf("CTB/CTB_128k_slack_5.h5", key="df", mode="r")
newDataset1_6 = pd.read_hdf("CTB/CTB_128k_slack_6.h5", key="df", mode="r")
dataset = pd.concat([
                            newDataset1_0,
                            newDataset1_1,
                            newDataset1_2,
                            newDataset1_3,
                            newDataset1_4,
                            newDataset1_5,
                            newDataset1_6,
                         ])
```

```python
test  = dataset[dataset['chromosome_name']=='chr8']
val   = dataset[dataset['chromosome_name']=='chr10']
train = dataset[dataset['chromosome_name']!='chr8'][dataset['chromosome_name']!='chr10']

len(test), len(val), len(train)
```

```python
X_trainpromoter         = np.array(list(train['sequence']))
y_train                 = train['labels'].values

X_validationpromoter    = np.array(list(val['sequence']))
y_validation            = val['labels'].values

X_testpromoter         = np.array(list(test['sequence']))
y_test                 = test['labels'].values
```

```python
k = 2**15
center = 128+2**16
leftpos  = center-k-1
rightpos = center+k-1
maxlen = rightpos-leftpos
leftpos, rightpos, maxlen
```

```python
X_train = tf.data.Dataset.from_tensor_slices(({'sequence': np.array(X_trainpromoter[:, leftpos:rightpos], dtype=np.uint8),      'target': y_train.reshape(-1,1)}))
X_val   = tf.data.Dataset.from_tensor_slices(({'sequence': np.array(X_validationpromoter[:, leftpos:rightpos], dtype=np.uint8), 'target': y_validation.reshape(-1,1)}))
X_test  = tf.data.Dataset.from_tensor_slices(({'sequence': np.array(X_testpromoter[:, leftpos:rightpos], dtype=np.uint8),       'target': y_test.reshape(-1,1)}))
cardinality = X_train.cardinality().numpy()
```

```python
batch_size      = 128 # 256
```

```python
X_train_tf = X_train.shuffle(cardinality*2).batch(batch_size).cache().prefetch(tf.data.AUTOTUNE).repeat()
X_val_tf   = X_val.batch(batch_size).cache().prefetch(tf.data.AUTOTUNE)
X_test_tf  = X_test.batch(batch_size).cache().prefetch(tf.data.AUTOTUNE)
```

# Train GX-BERT

```python
model_type      = "GX-BERT"
logdir          = None
loss            = "mse"
p_enc           = "original"
pooler          = "tanh"
n_layers1       = 0
n_layers2       = 1
k_size          = "multiconv_proj" # "multiconv"
shift_freq      = False
plot_attention  = True
alternate       = None
timeskip        = False
compression     = False
convDownScale   = True
tss             = None
mask_embedding  = 4
add_reg         = True
embed_dim       = 128
ff_dim          = embed_dim*4
pool_size       = 128
dense           = 256
lr              = 5e-5 # 1e-4
lr_red_epoch    = 30
reduction_factor= 5
t_rate          = 0.1    # 0.1
halflife        = False
transcr_factors = False
sure            = False
```

```python
# # Clear any logs from previous runs
# !rm -rf ./logs/transformer_best_0-001/
######################################################################################################################
import tensorflow as tf
tf.keras.backend.clear_session()

resolver = tf.distribute.cluster_resolver.TPUClusterResolver(tpu='grpc://' + os.environ['COLAB_TPU_ADDR'])
tf.config.experimental_connect_to_cluster(resolver)
# This is the TPU initialization code that has to be at the beginning.
tf.tpu.experimental.initialize_tpu_system(resolver)
print("All devices: ", tf.config.list_logical_devices('TPU'))

strategy = tf.distribute.TPUStrategy(resolver)

with strategy.scope():
    net = projTransformer(checkpoint_dir="", model_type=model_type, n_epochs=300, batch_size=batch_size, alternate=alternate, tss=tss, pool_size=pool_size, mask_embedding=mask_embedding,
                      learning_rate=lr, patience=30, optimizer="BERTAdam", vocab_size=5, k_size = k_size, pooler=pooler, shift_freq=shift_freq, timeskip=timeskip,
                      lr_reduction_epoch=lr_red_epoch, maxlen=maxlen, embed_dim=embed_dim, num_heads=4, ff_dim=ff_dim, dense=dense, plot_attention=plot_attention, add_reg=add_reg,
                      dropout_rate=0.1, logdir=logdir, t_rate=t_rate, momentum=0.9, loss=loss, n_layers1=n_layers1, n_layers2=n_layers2, compression=compression,
                      p_enc=p_enc, reduction_factor=reduction_factor)

########################################################################################################################
```

```python
X_train_d = strategy.experimental_distribute_dataset(X_train_tf)
X_val_d   = strategy.experimental_distribute_dataset(X_val_tf)
X_test_d  = strategy.experimental_distribute_dataset(X_test_tf)
```

The training phase takes Approximately one hour

```python
net.train_model(X_train_d,
                X_val_d,
                True,
                cardinality,
                strategy
                )
```

```python
net.evaluate_model(X_test_d, strategy=strategy)
print("")
predictions, y = net.evaluate_best_model(X_test_d, strategy=strategy)
print("")
net.plot_kde(X_test_d, strategy=strategy, predictions=predictions, y=y)
print("")
net.plot_r2(X_test_d, strategy=strategy, predictions=predictions, y=y)
print("")
```

```python
loss =  tf.keras.losses.mean_squared_error
test['predictions'] = predictions
test['loss'] = loss(test['labels'].values.reshape(-1,1), np.array(predictions).reshape(-1,1))
test
```

# Save Attention Tracks


```python
k=2**15
heads = 4
model_name = "GX-BERT"
for gene in test.index.values:
    print(gene)
    strand = test[test.index==gene].genomic_strand.values[0]
    chr    = test[test.index==gene].chromosome_name.values[0]
    tss    = test[test.index==gene].TSS.values[0]
    start  = tss - k
    end    = tss + k
    step   = 128
    iters  = int((end-start)/step)
    seq    = test[test.index==gene].sequence.values
    seq    = np.array(list(seq))
    seq    = np.array(seq[:, leftpos:rightpos], dtype=np.uint8)
    out    = net.model(seq)
    CLS    = out['Attention_Scores'][0].numpy()[:, :, 0, 1:]
    CLS_mean = np.mean(CLS, axis=1, keepdims=True)
    mean_att = np.mean(out['Attention_Scores'][0].numpy()[:, :, 1:, 1:], axis=(2, 1), keepdims=True)
    heads = 4
    if strand == "-":
        CLS           = np.flip(CLS, axis=2)
        CLS_mean      = np.flip(CLS_mean, axis=2)
        mean_att      = np.flip(mean_att, axis=3)

    model_name = "GX-BERT"
    if os.path.exists(f'BedTracks/{model_name}/{gene}/') == False:
        os.makedirs(f'BedTracks/{model_name}/{gene}/')
    for head in range(heads):
        with open(f'BedTracks/{model_name}/{gene}/{gene}CLS_head{head+1}.txt', 'w') as w_writer:
            w_writer.write(f"browser position {chr}:{start-k}-{end+k}\n\
browser hide all\n\
browser pack refGene encodeRegions\n\
browser full altGraph\n\
#	300 base wide bar graph, autoScale is on by default == graphing\n\
#	limits will dynamically change to always show full range of data\n\
#	in viewing window, priority = 20 positions this as the second graph\n\
#	Note, zero-relative, half-open coordinate system in use for bedGraph format\n\
track type=bedGraph name=\"{gene} {model_name} CLS Attention Head {head+1}\" description=\"{gene} {model_name} CLS Attention Head {head+1}\" visibility=full color=200,100,0 altColor=0,100,200 priority=20\n")
            for i in range(iters):
                w_writer.write(f'{chr} {start+i*step} {start+(i+1)*step-1} {CLS[0, head, i]}\n')

    with open(f'BedTracks/{model_name}/{gene}/{gene}CLS_mean.txt', 'w') as w_writer:
        w_writer.write(f"browser position {chr}:{start-k}-{end+k}\n\
browser hide all\n\
browser pack refGene encodeRegions\n\
browser full altGraph\n\
#	300 base wide bar graph, autoScale is on by default == graphing\n\
#	limits will dynamically change to always show full range of data\n\
#	in viewing window, priority = 20 positions this as the second graph\n\
#	Note, zero-relative, half-open coordinate system in use for bedGraph format\n\
track type=bedGraph name=\"{gene} {model_name} CLS Mean Attention\" description=\"{gene} {model_name} CLS Mean Attention\" visibility=full color=200,100,0 altColor=0,100,200 priority=20\n")
        for i in range(iters):
            w_writer.write(f'{chr} {start+i*step} {start+(i+1)*step-1} {CLS_mean[0, 0, i]}\n')
    with open(f'BedTracks/{model_name}/{gene}/{gene}mean_att.txt', 'w') as w_writer:
        w_writer.write(f"browser position {chr}:{start-k}-{end+k}\n\
browser hide all\n\
browser pack refGene encodeRegions\n\
browser full altGraph\n\
#	300 base wide bar graph, autoScale is on by default == graphing\n\
#	limits will dynamically change to always show full range of data\n\
#	in viewing window, priority = 20 positions this as the second graph\n\
#	Note, zero-relative, half-open coordinate system in use for bedGraph format\n\
track type=bedGraph name=\"{gene} {model_name} Mean Attention\" description=\"{gene} {model_name} Mean Attention\" visibility=full color=200,100,0 altColor=0,100,200 priority=20\n")
        for i in range(iters):
            w_writer.write(f'{chr} {start+i*step} {start+(i+1)*step-1} {mean_att[0, 0, 0, i]}\n')
```

# Save Gradient x Input Tracks (approx 10 minutes, approx 2GB)

```python
model_name = "GX-BERT"
for gene in test.index.values:
    print(gene)
    strand = test[test.index==gene].genomic_strand.values[0]
    loss =  tf.keras.losses.mean_squared_error
    chr   = test[test.index==gene].chromosome_name.values[0]
    tss   = test[test.index==gene].TSS.values[0]
    start = tss - k
    end   = tss + k
    step  = 128
    iters = int((end-start)/step)
    seq = test[test.index==gene].sequence.values
    seq = np.array(list(seq))
    seq = np.array(seq[:, leftpos:rightpos], dtype=np.uint8)

    label = test[test.index==gene].labels.values
    label = np.array([list(label)])
    label = np.array(label[:], dtype=np.float32)
    label, label.shape
    inputs = seq
    label = tf.convert_to_tensor(label, dtype=tf.float32)

    with tf.GradientTape() as tape:
        outputs     = net.model(inputs, training=False)['Regression_Output']
        vec_loss    = loss(label, outputs)
        loss        = tf.reduce_mean(vec_loss)

    grads = tape.gradient(loss, net.model.trainable_variables)
    seq_grads = grads[0].values.numpy()
    seq_grads.shape
    seq2 = np.array(seq, dtype=np.float32)
    w2v = net.model.layers[1](seq2)
    grad_x_input =  w2v[0]*seq_grads
    seq_grads_summed = np.abs(np.sum(grad_x_input[0], axis=1))
    step=1
    iters=seq_grads_summed.shape[0]
    if strand == "-":
        seq_grads_summed = np.flip(seq_grads_summed)

    if os.path.exists(f'BedTracks/{model_name}/{gene}/') == False:
        os.makedirs(f'BedTracks/{model_name}/{gene}/')
    with open(f'BedTracks/{model_name}/{gene}/{gene}gradient_x_input.txt', 'w') as w_writer:
        w_writer.write(f"browser position {chr}:{start-k}-{end+k}\n\
browser hide all\n\
browser pack refGene encodeRegions\n\
browser full altGraph\n\
#	300 base wide bar graph, autoScale is on by default == graphing\n\
#	limits will dynamically change to always show full range of data\n\
#	in viewing window, priority = 20 positions this as the second graph\n\
#	Note, zero-relative, half-open coordinate system in use for bedGraph format\n\
track type=bedGraph name=\"{gene} {model_name} Gradient x Input\" description=\"{gene} {model_name} Gradient x Input\" visibility=full color=200,100,0 altColor=0,100,200 priority=20\n")
        for i in range(iters):
            w_writer.write(f'{chr} {start+i*step} {start+(i+1)*step-1} {seq_grads_summed[i]}\n')
```
